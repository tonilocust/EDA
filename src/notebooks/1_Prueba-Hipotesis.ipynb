## ¿Test de hipótesis?

Una prueba o test de hipótesis es un **procedimiento estadístico** que se utiliza ***para tomar decisiones*** sobre una afirmación (o hipótesis) sobre una población. 

Esta afirmación se llama hipótesis nula y se contrasta con una hipótesis alternativa. La hipótesis nula generalmente afirma que no hay diferencia o efecto, mientras que la hipótesis alternativa afirma que hay una diferencia o efecto.

[Conceptos estadísticos previos](#conceptos-estadisticos)


## Procedimiento general:

#### 1. **Formulación de hipótesis:**


- Hipótesis nula (`H0`): Es la hipótesis que se quiere poner a prueba y se establece que no hay diferencia o relación entre las variables.

- Hipótesis alternativa (`H1`): Es la hipótesis opuesta a la hipótesis nula y se establece que hay una diferencia o relación entre las variables.

[Ejemplos de hipótesis](#ejemplos-de-hipotesis)

#### 2. **Selección del nivel de significancia:**

- El nivel de significancia (`α`) es el límite máximo de probabilidad de error que se está dispuesto a aceptar.

- Se suele elegir un nivel de significancia del 5% o 1%.

[Informacion adicional](#nivel-de-significancia)

#### 3. **Elección de la prueba estadística:**

- La [elección de la prueba estadística](#eleccion-de-la-prueba-estadistica) dependerá de las [características de los datos](2_Verificaciones-muestra.ipynb) y del tipo de hipótesis que se quiera probar.

- Pruebas estadísticas: 

    - **Test paramétricos**: aquellos que asumen una distribución específica de los datos y una igualdad en sus varianzas.

        - [T-test o t-Student](3_Test-parametricos.ipynb): se utiliza para comparar las medias de dos grupos independientes, cuando la variable a medir es continua y sigue una distribución normal.

        - [T-test pareada](3_Test-parametricos.ipynb): se utiliza para comparar las medias de dos grupos relacionados o apareados, cuando la variable a medir es continua y sigue una distribución normal.

        - [ANOVA](3_Test-parametricos.ipynb): se utiliza para comparar las medias de tres o más grupos independientes, cuando la variable a medir es continua y sigue una distribución normal.

        - ...

    - **Test NO paramétricos**: aquellos que no hacen supuestos sobre la distribución de los datos y la igualdad de varianzas.

        - [U-test](4_Test-no-parametricos.ipynb): se utiliza para comparar las medianas de dos grupos independientes, cuando la variable a medir es continua, pero no sigue una distribución normal.

        - [Prueba de Wilcoxon](4_Test-no-parametricos.ipynb): se utiliza para comparar las medianas de dos grupos relacionados o apareados, cuando la variable a medir es continua, pero no sigue una distribución normal.

        - [Prueba de Kruskal-Wallis](4_Test-no-parametricos.ipynb): Se utiliza para comparar las medianas de tres o más grupos independientes, cuando la variable a medir es continua, pero no sigue una distribución normal.

        - ...
        
#### 4. **Obtención del valor estadístico y el p-value (o valor p):**

- El *valor estadístico* se obtiene a través de la prueba estadística elegida.

- El `p-value` es la probabilidad de obtener un resultado igual o más extremo que el observado, asumiendo que la hipótesis nula es verdadera.

- Si el p-value es menor que el nivel de significancia elegido, se rechaza la hipótesis nula.

[Informacion adicional](#p-valor)

#### 5. **Toma de decisión:**

- `p-value < α`: 
    
    si el p-value es menor que el nivel de significancia elegido, **se rechaza la hipótesis nula** y se acepta la hipótesis alternativa.

- `p-value > α`: 
    
    si el p-value es mayor que el nivel de significancia elegido, **NO se puede rechazar la hipótesis nula** y se acepta que no hay suficiente evidencia para demostrar que la hipótesis alternativa es verdadera.

#### 6. **Interpretación de resultados:**

- Se debe interpretar los resultados en el contexto del problema y las limitaciones de la prueba estadística elegida.

- Se pueden realizar conclusiones y recomendaciones basadas en los resultados obtenidos.

## Conceptos estadisticos:

Para realizar un test de hipótesis, es necesario entender algunos **conceptos estadísticos** importantes, como:

1. **Muestra:** Es un subconjunto de la población que se selecciona para realizar una investigación.

2. **Población:** Es el conjunto total de individuos, objetos, eventos, etc., que se desea estudiar.

3. **Media:** Es el valor promedio de los datos en una muestra o población.

4. **Desviación estándar:** Es una medida de la variabilidad de los datos alrededor de la media.  Indica cuánto se alejan los datos individuales del valor promedio.

5. **Hipótesis nula:** Es una afirmación sobre la población que se asume que es verdadera hasta que se demuestre lo contrario.

6. **Hipótesis alternativa:** Es una afirmación que se propone para reemplazar a la hipótesis nula si se encuentra suficiente evidencia en contra de ella.

7. **Nivel de significancia:** Es la probabilidad de cometer un error de tipo I al rechazar la hipótesis nula cuando en realidad es verdadera.

8. **Estadístico de prueba:** Es un valor calculado a partir de los datos de la muestra, que se utiliza para decidir si se rechaza o no la hipótesis nula.

9. **Valor p:** Es la probabilidad de obtener un resultado igual o más extremo que el observado, suponiendo que la hipótesis nula es verdadera.

10. **Regla de decisión:** Es un conjunto de criterios preestablecidos para decidir si se rechaza o no la hipótesis nula en función del estadístico de prueba y el nivel de significancia elegidos.

Algunas de las **distribuciones más comunes** y sus características: 

- **Distribución uniforme:**

Cada valor en el intervalo de la distribución tiene la misma probabilidad de ocurrencia.
La función de densidad de probabilidad es constante dentro del intervalo y cero fuera de él.

Ejemplo: lanzamiento de un dado de seis caras.

- **Distribución binomial:**

Utilizada para modelar experimentos que tienen dos resultados posibles y mutuamente excluyentes.
La variable aleatoria de interés cuenta el número de éxitos en una cantidad fija de ensayos independientes y con idénticas probabilidades de éxito.
La función de densidad de probabilidad es la fórmula de la probabilidad binomial y su forma general es asimétrica.

Ejemplo: número de veces que una moneda cae en cara al lanzarla 10 veces.

- **Distribución de Poisson:**

Utilizada para modelar eventos raros y discretos que ocurren en un intervalo de tiempo o espacio fijo.
La variable aleatoria de interés cuenta el número de eventos en un intervalo específico.
La función de densidad de probabilidad es asimétrica y puede tomar valores discretos entre cero e infinito.

Ejemplo: número de clientes que llegan a una tienda en una hora determinada.

- **Distribución normal:**

Utilizada para modelar una gran variedad de fenómenos en diferentes campos.
La variable aleatoria de interés tiene una distribución simétrica en forma de campana.
La función de densidad de probabilidad tiene una fórmula matemática que depende de la media y la desviación estándar de la distribución.

Ejemplo: estaturas de los hombres adultos.

Las características de una **distribución normal**, también conocida como distribución de Gauss, son las siguientes:

- Simetría: la distribución normal es simétrica alrededor de su media. Esto significa que la probabilidad de que una variable aleatoria tome valores por encima de la media es igual a la probabilidad de que tome valores por debajo de la media.

- Campana: la distribución normal tiene una forma de campana, que se caracteriza por un pico en la media y una disminución gradual hacia los extremos. La altura del pico de la campana indica la probabilidad más alta de ocurrencia de la variable aleatoria.

- Unimodal: la distribución normal tiene un solo modo, lo que significa que hay un valor de la variable aleatoria que ocurre con mayor frecuencia que cualquier otro valor.
Asintótica: la curva de la distribución normal se acerca continuamente al eje horizontal, pero nunca lo alcanza. Esto significa que la probabilidad de que la variable aleatoria tome valores extremadamente grandes o extremadamente pequeños, aunque sea muy baja, nunca es cero.

- Media y desviación estándar: la distribución normal está completamente determinada por su media y su desviación estándar. La media es el centro de la distribución y la desviación estándar mide la dispersión de los valores de la variable aleatoria alrededor de la media.

- Teorema central del límite: la distribución normal es importante porque es la única distribución que satisface el teorema central del límite. Este teorema establece que, para una muestra lo suficientemente grande de una población, la distribución de las medias muestrales se aproxima a una distribución normal, independientemente de la distribución original de la población.

En resumen, la distribución normal es una distribución de probabilidad teórica que se utiliza para modelar muchos fenómenos naturales y humanos debido a sus características de simetría, campana, unimodalidad, asintoticidad y relación con la media y la desviación estándar.




## Ejemplos de hipotesis:



### Ejemplos de hipótesis válidas:

- **Hipótesis nula:** No hay diferencia significativa en la calificación promedio entre los estudiantes que asisten a clases presenciales y los que asisten a clases en línea.

- **Hipótesis alternativa:** La calificación promedio de los estudiantes que asisten a clases presenciales es mayor que la calificación promedio de los estudiantes que asisten a clases en línea.

En este ejemplo, las hipótesis están claramente definidas en términos de la población (estudiantes que asisten a clases presenciales y en línea) y la variable de interés (calificación promedio). La hipótesis nula establece que no hay diferencia significativa en la calificación promedio entre los dos grupos, mientras que la hipótesis alternativa establece que hay una diferencia significativa en favor del grupo de estudiantes que asisten a clases presenciales.

- **Hipótesis nula:** No hay relación entre el consumo de café y el riesgo de enfermedad cardíaca.

- **Hipótesis alternativa:** El consumo de café se asocia con un mayor riesgo de enfermedad cardíaca.

En este ejemplo, las hipótesis están claramente definidas en términos de la población (personas que consumen café) y la variable de interés (riesgo de enfermedad cardíaca). La hipótesis nula establece que no hay relación entre las dos variables, mientras que la hipótesis alternativa establece que hay una relación entre ellas.



### Ejemplos de hipótesis no válidas:

- **Hipótesis nula:** El clima es agradable.

- **Hipótesis alternativa:** El clima es desagradable.

En este ejemplo, las hipótesis no están claramente definidas en términos de la población y la variable de interés. Además, no se establece una diferencia clara entre la hipótesis nula y la alternativa.

- **Hipótesis nula:** Los gatos son mejores que los perros.

- **Hipótesis alternativa:** Los perros son mejores que los gatos.

En este ejemplo, las hipótesis no están claramente definidas en términos de la población y la variable de interés. Además, estas hipótesis no son adecuadas para una prueba de hipótesis ya que no se pueden medir objetivamente y son subjetivas.

## Nivel de significancia: 

El nivel de significancia es una **medida de la probabilidad de cometer un error tipo I** (rechazar la hipótesis nula cuando es verdadera) en una prueba de hipótesis. Por lo tanto, es importante seleccionar un nivel de significancia apropiado antes de realizar la prueba de hipótesis.

</div>
<br>

> El nivel de significancia más *comúnmente utilizado es 0.05*, lo que significa que hay una probabilidad del 5% de rechazar la hipótesis nula cuando es verdadera. Sin embargo, el nivel de significancia también puede ser seleccionado en función del contexto de la investigación o el experimento, así como de la importancia de cometer un error tipo I o tipo II.

Para seleccionar el nivel de significancia, se debe tener en cuenta la naturaleza de la hipótesis, la cantidad de datos disponibles y la importancia práctica de los resultados de la prueba de hipótesis. 

Por ejemplo, en una prueba de hipótesis para evaluar la seguridad de un nuevo medicamento, un nivel de significancia más bajo (por ejemplo, 0.01) podría ser apropiado para minimizar el riesgo de cometer un error tipo I y proteger la salud de los pacientes.

## P-valor:

El valor p (p-value) es una **medida estadística utilizada en pruebas de hipótesis** para determinar la probabilidad de obtener un resultado igual o más extremo que el observado, asumiendo que la hipótesis nula es verdadera. En otras palabras, `el valor p indica la evidencia en contra de la hipótesis nula proporcionada por los datos muestrales`.

El valor p se expresa como un número entre 0 y 1, y se interpreta de la siguiente manera: si el valor p es menor que el nivel de significancia previamente seleccionado, se rechaza la hipótesis nula en favor de la hipótesis alternativa. Por ejemplo, si el nivel de significancia es 0.05 y el valor p es 0.02, se rechaza la hipótesis nula porque el valor p es menor que el nivel de significancia. Esto significa que hay suficiente evidencia estadística para apoyar la hipótesis alternativa.

Por otro lado, si el valor p es mayor que el nivel de significancia, no se rechaza la hipótesis nula. Por ejemplo, si el nivel de significancia es 0.05 y el valor p es 0.10, no se rechaza la hipótesis nula porque el valor p es mayor que el nivel de significancia. Esto significa que no hay suficiente evidencia estadística para rechazar la hipótesis nula.

Es importante tener en cuenta que el valor p no indica la probabilidad de que la hipótesis nula sea verdadera o falsa. Solo indica la evidencia proporcionada por los datos muestrales en contra de la hipótesis nula. Además, el valor p no proporciona información sobre la magnitud o importancia de la diferencia observada entre las dos muestras o poblaciones. Por lo tanto, es importante interpretar los resultados de una prueba de hipótesis con precaución y considerar la relevancia práctica de los resultados obtenidos.

El cálculo del valor p depende del tipo de prueba de hipótesis que se esté realizando y del tipo de estadística utilizada. En general, el valor p se calcula a partir de la distribución de probabilidad correspondiente al estadístico de prueba utilizado.

## Eleccion de la prueba estadistica: 

La selección de la prueba estadística a utilizar en un test de hipótesis depende en gran medida del tipo de datos que se tienen y de la pregunta que se quiere responder. 

En general, existen **dos tipos principales de datos**: variables `cuantitativas` y variables `cualitativas`.


### Para variables cuantitativas:

**Verificaciones:**
1. *Verificar la normalidad*: Lo primero que se debe hacer es verificar si los datos siguen una distribución normal. Esto se puede hacer a través de pruebas estadísticas como la prueba de normalidad de Kolmogorov-Smirnov o la prueba de Shapiro-Wilk.

2. *Variabilidad de las muestras*: Es importante verificar que la varianza de las muestras sea igual. Esto se puede hacer utilizando la prueba de igualdad de varianzas como la prueba de Levene o la prueba de Bartlett.

**Selección de la prueba estadística:**

Una vez verificada la normalidad y la igualdad de varianzas, se puede seleccionar la prueba estadística adecuada. 

- Si se está comparando la `media de dos muestras`, se puede utilizar el `t-test` de dos muestras independientes (si las muestras no están relacionadas) o el `t-test pareado` de dos muestras (si las muestras están relacionadas). 

- Si se quiere comparar la `media de más de dos muestras`, se puede utilizar el análisis de varianza (`ANOVA`). 

- Si se desea establecer una relación entre dos variables, se puede utilizar la `regresión lineal`.


### Para variables cualitativas:

**Verificaciones:**

- *Tipo de variable:* es importante determinar si las variables son `nominales` o `ordinales`. 

    - Las variables *nominales* son aquellas que no tienen un orden lógico, como el color de ojos o la preferencia por un deporte en particular. 

    - Las variables *ordinales* son aquellas que tienen un orden lógico, como la calificación de un producto en una escala de 1 a 5.

**Selección de la prueba estadística:**

- Si se están comparando dos grupos de datos categóricos, se puede utilizar la prueba de `chi-cuadrado`. 

- Si se quiere determinar si hay una asociación entre dos variables categóricas, se puede utilizar la `prueba exacta de Fisher`. 

- Si se tiene una variable ordinal y una nominal, se puede utilizar la prueba de `Mann-Whitney` o la prueba de `Kruskal-Wallis`. 

- Si se tiene una variable ordinal y una variable cuantitativa, se puede utilizar la `regresión ordinal`.
